```python
import albumentations as albu
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = 'C:\\Users\\gueganj\\Desktop\\My_DataBase\\nature\\\images\\013_scanRecord_4.jpg'
mask_path  = 'C:\\Users\\gueganj\\Desktop\\My_DataBase\\nature\\masks\\frame\\013_scanRecord_4.png'

image = plt.imread(image_path)
mask  = plt.imread(mask_path)
```

```python
def display_result(augmentation, image, multiple=False):
    if not multiple:
        augmented = augmentation(image=image)
        plt.imshow(augmented['image'])
        plt.show()
    else:
        plt.figure(figsize=(15,5))
        for i in range(10):
            plt.subplot(2,5,i+1)
            augmented = augmentation(image=image)
            plt.imshow(augmented['image'])
            plt.xticks(()),plt.yticks(())
        plt.tight_layout()
        plt.show()
```

# HorizontalFlip

```python
augmentation = albu.HorizontalFlip(p=1)
display_result(augmentation, image)
```

# ShiftScaleRotate

```python
augmentation = albu.ShiftScaleRotate(scale_limit=(-0.5, 0.5), rotate_limit=[-15,15], shift_limit=[-0.1,0.1], p=1, border_mode=0)
display_result(augmentation, image, multiple=True)
```

# IAAAdditiveGaussianNoise

```python
augmentation = albu.IAAAdditiveGaussianNoise(p=1)
display_result(augmentation, image)
```

# GaussNoise

```python
augmentation = albu.GaussNoise(p=1)
display_result(augmentation, image)
```

# ISOnoise

```python
augmentation = albu.ISONoise(p=1)
display_result(augmentation, image)
```

# IAAPerspective

```python
augmentation = albu.IAAPerspective(p=1)
display_result(augmentation, image, multiple=True)
```

# Luminosity

```python
augmentation = albu.CLAHE(p=1)
display_result(augmentation, image, multiple=True)
```

```python
augmentation = albu.RandomBrightness(p=1)
display_result(augmentation, image, multiple=True)
```

```python
augmentation = albu.RandomGamma(p=1)
display_result(augmentation, image, multiple=True)
```

# Blur

```python
augmentation = albu.IAASharpen(p=1)
display_result(augmentation, image, multiple=True)
```

```python
augmentation = albu.Blur(blur_limit=15, p=1)
display_result(augmentation, image, multiple=True)
```

```python
augmentation = albu.MotionBlur(blur_limit=30, p=1)
display_result(augmentation, image, multiple=True)
```

# Contrast

```python
augmentation = albu.RandomContrast(p=1)
display_result(augmentation, image, multiple=True)
```

```python
augmentation = albu.HueSaturationValue(p=1)
display_result(augmentation, image, multiple=True)
```

# Distortion

```python
augmentation = albu.OpticalDistortion(p=1, distort_limit=0.5)
display_result(augmentation, image, multiple=True)
```

```python
augmentation = albu.GridDistortion(p=1,distort_limit=0.2)
display_result(augmentation, image, multiple=True)
```

```python
augmentation = albu.IAAPiecewiseAffine(p=1, scale=(0.03, 0.03))
display_result(augmentation, image, multiple=True)
```

# RGBshift

```python
augmentation = albu.RGBShift(p=1)
display_result(augmentation, image, multiple=True)
```
